iloc[] (Integer-based selection):
Selects rows and columns based on integer positions (0-based indices).
Selects rows and columns based on integer positions (0-based indices).
Does not accept boolean data.
Exclusive of the last element in the range.
Raises an index error if the position doesn’t exist.



loc[] (Label-based selection):
Selects rows and columns based on labels (row and column names).
Accepts boolean data.
Inclusive of the last element in the range.
Raises a key error if the label is not present.

In Pandas, the mode() method returns the mode value for each column in a DataFrame. The mode represents the most common value(s) in the data. It can be multiple values if there are ties.
